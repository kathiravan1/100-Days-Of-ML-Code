```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder

# Read the training, testing, and validation datasets from CSV files
train_df = pd.read_csv('/content/drive/MyDrive/SharedTask/ta-en-misogyny-train.csv',delimiter='\t', header=None, names=['word', 'tag'],usecols=[0, 1], quoting=3)
test_df = pd.read_csv('/content/drive/MyDrive/SharedTask/ta-en-misogyny-dev.csv',delimiter='\t', header=None, names=['word', 'tag'],usecols=[0, 1], quoting=3)
validation_df = pd.read_csv('/content/drive/MyDrive/SharedTask/ta-en_test_without_labels.csv', header=None, names=['tag'],usecols=[0])


```

```python
train_df.head(5)
```

```python
# Assuming your DataFrame is named 'df' and the column with double quotes is named 'column_name'
train_df['word'] = train_df['word'].str.replace('"', '')
```

```python
test_df.head(5)
```

```python
# Extract the text data and labels from the datasets
train_data = train_df['tag'].tolist()
train_labels = train_df['word'].tolist()
test_data = test_df['tag'].tolist()
test_labels = test_df['word'].tolist()
validation_data = validation_df['tag'].tolist()
```

```python
# Preprocess test data to handle missing values
test_data = ["" if pd.isnull(tag) else tag for tag in test_data]
```

```python

# Instantiate the vectorizer
vectorizer = CountVectorizer()

# Fit the vectorizer with the training data
vectorizer.fit(train_data)

# Transform the training data
vectorized_train_data = vectorizer.transform(train_data)

# Transform the testing data
vectorized_test_data = vectorizer.transform(test_data)

# Transform the validation data
vectorized_validation_data = vectorizer.transform(validation_data)
```

```python
# Encode the string labels to numerical values
label_encoder = LabelEncoder()
encoded_train_labels = label_encoder.fit_transform(train_labels)
encoded_test_labels = label_encoder.transform(test_labels)
```

```python
# Instantiate the SVM classifier
svm_model = svm.SVC()

# Train the SVM model
svm_model.fit(vectorized_train_data, encoded_train_labels)

# Make predictions on the testing data
test_predictions = svm_model.predict(vectorized_test_data)
```

```python
# Calculate macro-averaged Precision, Recall, and F-Score on the testing data
precision_macro_avg = precision_score(encoded_test_labels, test_predictions, average='macro')
recall_macro_avg = recall_score(encoded_test_labels, test_predictions, average='macro')
f1_macro_avg = f1_score(encoded_test_labels, test_predictions, average='macro')
```

```python
# Print the results on testing data
print("Testing Data - Macro-Averaged Precision:", precision_macro_avg)
print("Testing Data - Macro-Averaged Recall:", recall_macro_avg)
print("Testing Data - Macro-Averaged F-Score:", f1_macro_avg)
```

```python
# Make predictions on the validation data
#validation_predictions = svm_model.predict(vectorized_validation_data)
```

```python
# Make predictions on the validation data
validation_predictions = svm_model.predict(vectorized_validation_data)

# Decode the numerical predictions back to string labels
decoded_validation_predictions = label_encoder.inverse_transform(validation_predictions)

# Print the predictions on the validation data
for i, prediction in enumerate(decoded_validation_predictions):
    print("Validation Example:", validation_data[i])
    print("Prediction:", prediction)
    print()
```

```python
# Calculate macro-averaged F-Score on the validation data
f1_macro_avg_val = f1_score(validation_labels, validation_predictions, average='macro')

# Print the macro-averaged F-Score on the validation data
print("Validation Data - Macro-Averaged F-Score:", f1_macro_avg_val)
```
